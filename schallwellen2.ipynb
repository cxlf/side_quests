```julia
using Plots
plotlyjs()
```

```julia
c = 340
f = 60:20000
lambda = c ./ f
```

```julia
plot(f, lambda, xlabel="Frequenz (Hz)", ylabel="Wellenlänge (m)", title="Wellenlängen in Abhängigkeit von der Frequenz")
```

```julia
d = sqrt(3.4^2+25) - sqrt(1.6^2+25) #sqrt(3.4^2+25) ~= 6.0m ist die entfernung lautsprecher 1 und sqrt(1.6^2+25) ~= 5.25m ist die entfernung von lautsprecher 2

k = 0:0.5:10
```

```julia
konstr_inteferenz, destr_inteferenz = c ./ f

konstr_inteferenz = d ./ k                  #(k .- 0.5) 
destr_inteferenz = d ./ (k .+ 0.5)

plot(k, konstr_inteferenz, label="Konstruktive Interferenz", xlabel="k. Ordnung", ylabel="Wellenlänge (m)")
scatter!(k, konstr_inteferenz)

plot!(k, destr_inteferenz, label="Destruktive Interferenz")
scatter!(k, destr_inteferenz)
```

```julia
c ./ f = d ./ k
```
